```python
import csv
import sys

class TopCounts(object):
    """
    Computes top counts for a read in CSV file 
    
    Args:
        input_filename (str): CSV filename where the data is located 
        output_filename (str): filename where to save output 
        delimiter (str): CSV delimiter character 
        header (boolean): If the CSV file contains a header
        quotechar (str): Quote character in CSV file 

    Attributes:
        input_filename (str): CSV filename
        output_filename (str): output filename
        delimiter (str): delimiter character for read and write  
        header (boolean): If the CSV file contains a header
        quotechar (str): Quote character
    """
    def __init__(self, input_filename, output_filename, header=True, delimiter=';', quotechar='"'):
        self.input_filename = input_filename
        self.output_filename = output_filename
        self.delimiter = delimiter
        self.header = header
        self.quotechar = quotechar
        
    def compute_counts(self, status_col_name='CASE_STATUS', occupation_col_name='SOC_NAME', 
                       certified_col_name='CERTIFIED'):
        """
        Computes counts for desired column field based on column names 
        Args:
            param1 (int): The first parameter.
            param2 (:obj:`str`, optional): The second parameter. Defaults to None.
                Second line of description should be indented.
            *args: Variable length argument list.
            **kwargs: Arbitrary keyword arguments.

        Returns:
            counts (dict): dictionary of {desired_col : count}
            N (int): total number of certified applications regardless of occupation
        """
        counts = {}
        with open(self.input_filename, 'r') as f:
            # read file 
            lines = csv.reader(f, quotechar=self.quotechar, delimiter=self.delimiter,
                             quoting=csv.QUOTE_ALL)
            if self.header:
                # get column names
                columns = next(lines)
                print(columns)
                status_idx = columns.index(status_col_name)
                print(status_idx)
                occupation_col_idx = columns.index(occupation_col_name)
                print(occupation_col_idx)
            else:
                status_idx = sys.args()
                occupation_col_idx = sys.args()
            # get total number of certified applications regardless of occupation
            N = 0
            for line in lines:
                if line[status_idx] == certified_col_name:
                    desired = line[occupation_col_idx]
                    N += 1
                    # increment count based on occupation
                    if desired not in counts:
                        counts[desired] = 0
                    counts[desired] += 1
        # return count dict and total number of certified applications regardless of occupation
        return counts, N


    def sort_values(self, counts, top_N=10):
        """
        Sorts `counts` to return `top_N` columns and counts 
        Args:
            counts (dict): Count dictionary
            top_N (int): Number of top columns 
        
        Returns:
            top (list): top N list of tuples [(desired_col, counts), ...]
        """
        top = sorted(counts.items(), key=lambda x: x[1], reverse=True)[:top_N]
        return top
    
    def add_percentage(self, top, N):
        """
        Add percentage of applications that have been certified compared 
        to total number of certified applications regardless of state. 
        Args:
            top (list): top N list of tuples [(desired_col, counts), ...]
            N (int): total number of certified applications regardless of occupation
        
        Returns:
            (list): top N list of tuples [(desired_col, counts, percentage), ...]
        """
        print([top[i] + ((((round(top[i][1] / N, 3)) * 100.0),)) for i in range(len(top))])
        return [top[i] + ((((round(top[i][1] / N, 3)) * 100.0),)) for i in range(len(top))]
    
    def write_to_file(self, top):
        """
        Saves output to output_filename
        Args:
            top (list): top N list of tuples [(desired_col, counts), ...]
        """
        with open(self.output_filename, 'w') as f:
            f.write('TOP_OCCUPATIONS' + self.delimiter + 'NUMBER_CERTIFIED_APPLICATIONS'  + self.delimiter + 'PERCENTAGE' + '\n')
            for line in top:
                line = self.delimiter.join(map(str, line))
                f.write(line + '\n')
        

def main():
    occupation_top = TopCounts(input_filename='../input/H1B_FY_2016.csv', 
                               output_filename='../output/top_10_occupations.txt', 
                               header=True, delimiter=';')
    counts, N = occupation_top.compute_counts(status_col_name='CASE_STATUS', occupation_col_name='SOC_NAME', 
                       certified_col_name='CERTIFIED')
#     counts, N = occupation_top.compute_counts(status_col_name='STATUS', occupation_col_name='LCA_CASE_SOC_NAME', 
#                        certified_col_name='CERTIFIED')
    top = occupation_top.sort_values(counts, top_N=10)
    top = occupation_top.add_percentage(top, N)
    occupation_top.write_to_file(top)

if __name__ == '__main__':
    main()
```
